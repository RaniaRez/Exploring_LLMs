```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

### Load the Mistral7B model
This would require a GPU instance

```python
#Load the model from its source
model_name= "TheBloke/Mistral-7B-Instruct-v0.1-GPTQ"
model = AutoModelForCausalLM.from_pretrained(model_name,
                                             device_map="auto",
                                             trust_remote_code=False, #for the sake of security
                                             revision="main")
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python
#defines the prompt to use for text generation

prompt = "Predict the impact of artificial intelligence on human learning. write the output in the form of a timeline marking the dates"
#The template of the prompt
prompt_template=f'''<s>[INST] {prompt} [/INST]
'''
#creates a text generation pipeline using the Hugging Face Transformers
pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    do_sample=True, #making the output more diverse by generating text in a sampling-based manner
    temperature=0.7,#randomness of the output
    top_p=0.95, #considering the top p values
    top_k=40,
    repetition_penalty=1.1
)
```

```python
#generate text based on the prompt_template and prints that generated text
tok = tokenizer("prompt_template")
tokens = len(tok['input_ids'])
print(f"Number of tokens: {tokens}")
import time
start_time = time.time()
print(pipe(prompt_template)[0]['generated_text'])
end_time = time.time()
execution_time_mistral = end_time - start_time
print(f"Generated in: {execution_time_mistral} seconds")
```
